```python
import pygame, sys, random #pygame: thu vien ve, xu ly phim, am thanh,sys dung de sys.exits() khi thoat
pygame.init() #Khoi tao cac module can thiet cua Pygame

#Thet lap cua so
WIDTH, HEIGHT = 960, 540 #kich thuoc cua so
screen = pygame.display.set_mode((WIDTH, HEIGHT)) #screen: be mat chinh de ve mo thu len
pygame.display.set_caption("Trap Adventure - Mini Edition") #tieu de cua so

#Tai hinh nen
sky = pygame.image.load("sky.jpg").convert() #Tai anh va chuan hoa format de ve nhanh hon
ground = pygame.image.load("ground.png").convert()
sky = pygame.transform.scale(sky, (WIDTH, HEIGHT)) #Thay doi kich thuoc anh de khop cua so
ground_height = 120 #chieu cao vung dat (120px): ve ground o cuoi man hinh bang HEIGHT - ground_height
ground = pygame.transform.scale(ground, (WIDTH, ground_height))

#Mau sac
PLAYER_COLOR = (80, 200, 255)
TRAP_COLOR = (255, 60, 60)

#Setup nhan vat
player = pygame.Rect(100, HEIGHT - ground_height - 70, 40, 60) #Hinh chu nhat, vi tri khoi tao dung tren mat dat
player_speed = 5 #Toc do dich chuyen ngang (pixel/frame)
velocity_y = 0 #Van toc theo truc y (duong xuong, am len)
gravity = 0.8 #Do tang van toc y moi frame (mo phong trong luc)
is_jumping = False #co de ko cho nhay lien tuc khi dang o giua khong trung
dead = False

# Traps 
traps = [
    pygame.Rect(400, HEIGHT - ground_height - 20, 40, 20),
    pygame.Rect(700, HEIGHT - ground_height - 80, 40, 80)
]

#Phong chu va dong ho
font = pygame.font.SysFont("Consolas", 48)
clock = pygame.time.Clock() #kiem soat FPS (do muot)

def draw_game():
    screen.blit(sky, (0, 0)) #screen.blit(image, pops): ve image len screen theo thu tu: nen-ground-trap-player
    screen.blit(ground, (0, HEIGHT - ground_height))
    for trap in traps:
        pygame.draw.rect(screen, TRAP_COLOR, trap) #pygame.draw.rect(...): ve hinh chu nhat (trap va player neu muon fallback khong cn anh)
    pygame.draw.rect(screen, PLAYER_COLOR, player, border_radius=6)

def reset_game():
    global player, velocity_y, dead #global: de thay doi toan cuc trong ham
    player.x, player.y = 100, HEIGHT - ground_height - 70
    velocity_y = 0 #Dat lai vi tri nhan vat doc ve 0
    dead = False #bat dau choi lai

# --- GAME LOOP ---
while True:
    for event in pygame.event.get(): #lay danh sach su kien (nhan phim, dong cua so,...)
        if event.type == pygame.QUIT: #Neu nguoi dung dong cua so->goi pygame.quit()->sys.exit() de thoat sach
            pygame.quit()
            sys.exit()
#Lay trang thai phim
    keys = pygame.key.get_pressed() #keys: mang boolean cho biet phim nao dang duoc giu xuong (True/False)
#Cap nhat khi chua die
    if not dead:
        #Di chuyen ngang
        if keys[pygame.K_LEFT]: #Dich player.x trai/phai theo player_speed. Khong co gioi han man (co the di ra ngoai cua so neu khong chan)
            player.x -= player_speed
        if keys[pygame.K_RIGHT]:
            player.x += player_speed
        #Nhay
        if keys[pygame.K_SPACE] and not is_jumping:
            velocity_y = -15
            is_jumping = True #Khi giu space ma is_jumping la False (dang dung tren nen), gan van toc doc am (len tren) de nhay. 
#Sau do dat is_jumping = True de chan nhay kep

        #Trong luc va cap nhat y: Cap nhat vi tri theo velocity_y
        player.y += velocity_y #tang velocity_y theo gravity -> tao parabol nhay
        velocity_y += gravity

        #Va cham san
        if player.bottom >= HEIGHT - ground_height:
            player.bottom = HEIGHT - ground_height
            velocity_y = 0
            is_jumping = False
#Neu day player cham or vuot qua mep tren cua ground thi gan lai cho chinh xac (dat dung tren mat dat)
#dat van toc doc =0 va bat lai kha nang nhay (is_jumping = False)
        #Kiem tra traps
        for trap in traps:
            if player.colliderect(trap):
                dead = True #True neu 2 Rect giao nhau (va cham)
                break

    else:
        # Death screen
        text = font.render("YOU DIED", True, (255, 0, 0))
        screen.blit(sky, (0, 0))
        screen.blit(text, (WIDTH // 2 - 150, HEIGHT // 2 - 50))
        pygame.display.flip() #Day frame len man hinh
        pygame.time.wait(1500) #Tam dung chuong trinh 1,5s (1500ms) de nguoi choi xem thong bao
 #time.wait() block toan bo vong lap (khong xu ly su kien), neu nguoi choi co dong cua so trong luc cho co the khong thay phan hoi       
        reset_game()

    #Ve va gioi han FPS
    draw_game() #Ve moi thu
    pygame.display.flip() #Cap nhat man hinh
    clock.tick(60) #Gioi han vong lap chay toi da 60 lan/giay (FPS = 60), giup game muot va dong deu tren may khac nhau
```
