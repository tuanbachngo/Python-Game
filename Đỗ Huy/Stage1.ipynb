```python
import pygame
pygame.init()
WIDTH, HEIGHT = 960, 540
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Hiep si 36")
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
pygame.quit()

#################################################################

FPS = 60
TILE = 48

# Vật lý đơn giản
GRAVITY = 0.6
MAX_FALL = 14
PLAYER_SPEED = 4.2
JUMP_VEL = -12
COYOTE_MS = 120  # nhảy "cứu cánh" sau khi rời đất ~0.12s

# Màu sắc
BG = (20, 24, 28)
WHITE = (240, 240, 240)
GREY = (120, 128, 140)
GREEN = (80, 200, 120)
RED = (220, 60, 80)
YELLOW = (250, 210, 90)
CYAN = (100, 210, 230)
BROWN = (120, 90, 60)
ORANGE = (240, 150, 60)

#################################################################

class player:
    def __init__(self, x, y): #x, y là tọa độ nhân vật/object khi class này được gọi
        self.rect = pygame.Rect(x, y, 20, 30) #hình chữ nhật đại diện cho nhân vật/object, 20x30 là kích thước
        self.vel_x = 0
        self.vel_y = 0
        self.is_on_ground = False
        self.heart = 1

class spikes: #active_after: thời gian chờ trước khi gai hoạt động (ms) nếu muốn gai mọc lên sau 1 khoảng tgian và có animation, có biến này nếu gai là loại active khi interact
    def __init__(self, x, y, active_after = 200):
        self.rect = pygame.Rect(x, y, TILE, TILE)
        self.damage = 1
        self.active_after = active_after

class platform:
    def __init__(self, x, y, w, h):
        self.rect = pygame.Rect(x, y, w, h)
        self.triggered = False #biến này dùng để kiểm tra xem người chơi đã chạm vào platform chưa (dùng cho platform di chuyển)
        self.moving = False #state di chuyển của platform

class moving_snails:
    def __init__(self, x, y, path_length, speed):
        self.rect = pygame.Rect(x, y, TILE, TILE)
        self.path_length = path_length #quãng đường di chuyển
        self.speed = speed #tốc độ
        self.direction = 1 #phương hướng của snail
        self.start_x = x
        self.damage = 0.5 #để người chơi chạm được 1 lần vào snail, chỉ có 1 cơ hội nhảy, tùy sau này code level thì gtri này sửa được

class check_point:
    def __init__(self, x, y):
        self.rect = pygame.Rect(x, y, TILE, TILE)
        self.activated = False #check xem checkpoint hđ chưa, tuần sau viết 1 hàm thay đổi state của checkpoint

class firethrower:
    def __init__(self, x, y, direction, active_after = 300):
        self.rect = pygame.Rect(x, y, TILE, TILE)
        self.direction = direction #hướng bắn lửa: 'left', 'right', 'up', 'down', tùy hướng thì hàm spawn ra object sẽ load animation khác nhau
        self.active_after = active_after
        self.damage = 1

class rolling_stone:
    def __init__(self, x, y, speed):
        self.rect = pygame.Rect(x, y, TILE, TILE)
        self.speed = speed
        self.damage = 1

class world: #world sẽ chứa tất cả các object trong level
    def __init__(self):
        self.platforms = []
        self.spikes = []
        self.moving_snails = []
        self.check_points = []
        self.firethrowers = []
```
